1. Вы открываете бутылку молока и делаете из нее глоток. Вероятность того, что молоко скисло - 0.02. С какой вероятностью молоко окажется свежим?

События:

A: молоко скисло;
B: молоко оказалось свежим.

P(A)+P(B)=1

P(B)=1-P(A)=1-0,02=0,98

2. На пешеходном переходе установлен светофор с 2 лампами. Вероятность перегорания лампы в течение года - 0.2. С какой вероятностью не перегорит хотя бы одна лампа?

1-й вариант

События:
А: первая лампа перегорит в течение года; и
B: вторая лампа перегорит в течение года.

P(A)=P(B)=0,2

C=AxB - перегорели обе лампы в течение года;

P(C)=P(AxB)=P(A)xP(B)=0,2x0,2=0,04

C̄ - перегорела хотя бы одна лампа в течение года

P(C̄) = 1-P(C)=1-0,04=0,96


2-й вариант

События:
A: в течение года не перегорит первая лампа; или
B: в течение года не перегорит вторая лампа.

P(A+B)=P(A)+P(B)-P(AxB)

P(A)=P(B)=1-0,2=0,8

P(AxB)=0,8x0,8=0,64

P(A+B)=0,8+0,8-0,64=0,96


3. Менеджер по продажам совершает сделку если знаком с клиентом с вероятностью - 0.8, а если ничего о нем не слышал с вероятностью 0.1. Начальник передает менеджеру список из 10 клиентов, 4 из которых менеджеру известны. С какой вероятностью выбрав случайного клиента из списка менеджер ничего ему не продаст?

События:

A: менеджер совершает сделку с клиентом, которого знает;
B: менеджер совершает сделку с клиентом, которого не знает;

С: менеджер совершит сделку;

надо найти вероятность события C̄, при котором менеджер ничего не продаст клиенту.

P=m/n, где

n=10;
m-число благоприятных исходов.

A: вероятность того, что клиент знаком менеджеру - 2/5 = 0,4;
   менеджер совершает сделку с клиентом, которого знает - 0,8
   
P(A)=0,4x0,8=0,32

B: вероятность того, что менеджер клиента не знает - P(B)=1-P(A)=(1-0,4) = 0,6;
   менеджер совершает сделку с клиентом, которого не знает - 0,1

P(B)=0,6x0,1=0,06

P(C)=P(A+B)=P(A)+P(B)=0,32+0,06=0,38

P(C̄)=1-P(C)=1-0,38=0,62

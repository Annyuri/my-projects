```python
# 15. У какого категориального признака наблюдается самый сильный диcбаланс классов?

Obj = data[['sex', 'chest_pain_type', 'fasting_blood_sugar', 'rest_ecg', 'exercise_induced_angina', 'st_slope', 'thalassemia']].describe().T
Obj
```

```python
df=pd.DataFrame(Obj,columns=["count", "freq", "ratio"])
df
```

```python
df=pd.DataFrame(Obj,columns=["count", "freq", "ratio"])
values=df['freq']/df['count']*100
df["ratio"]=values
df
```

```python
df['ratio'].sort_values(ascending=False)
```

Ответ: у категориального признака fasting_blood_sugar наблюдается самый сильный дисбаланс классов
